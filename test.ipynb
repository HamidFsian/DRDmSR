```python
import numpy as np
import matplotlib.pyplot as plt
import cv2


# Load your data using np.load
target_demosaic = np.load("estimated_demosaic.npy")  # Make sure to provide the correct file path
print(target_demosaic.shape)
mean = np.mean(target_demosaic)
std = np.std(target_demosaic)
img_normalized = (target_demosaic - mean) / std

img_normalized = target_demosaic * 255.
img_normalized[img_normalized < 0] = 0
img_normalized[img_normalized > 255.] = 255.


# print(np.max(img_normalized))
# print(np.min(img_normalized))

print(np.max(img_normalized))
print(np.min(img_normalized))

# Plot the data using plt.imshow
plt.imshow(img_normalized[0,:,:], cmap='gray')  # Adjust the colormap ('viridis' is just an example)
plt.colorbar()  # Add a colorbar for reference
plt.title("Raw Data")  # Set the plot title
plt.xlabel("X-axis")  # Label the x-axis (customize as needed)
plt.ylabel("Y-axis")  # Label the y-axis (customize as needed)
plt.show()  # Display the plot
```
